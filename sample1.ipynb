```python
# 正規化 1: z-score = (data - 平均值) / 標準差
import numpy as np
def zscore(x, axis=None):
    xmean = x.mean(axis=axis, keepdims=True)
    xstd = np.std(x, axis=axis, keepdims=True)
    zscore = (x - xmean) / xstd
    return zscore

a = np.random.randint(10, size=(2, 5))
print(a)
print(zscore(a))
print(zscore(a, axis=1))
```

```python
# 正規化 2: min-max = (data - 最小值) / (最大值 - 最小值)
def min_max(x, axis=None):
    min = x.min(axis=axis, keepdims=True)
    max = x.max(axis=axis, keepdims=True)
    result = (x - min) / (max - min)
    return result

b = np.random.randint(10, size=(2, 5))
print(b)
print(min_max(b))
print(min_max(b, axis=1))
```

```python
# 迴歸分析
import matplotlib.pyplot as plt
# 建立假資料: 20個點
np.random.seed(7)

x = np.random.rand(20) * 8 - 4
y = np.sin(x) + np.random.rand(20) * 0.2

# 喂資料給機器學習, 求逼近20個點的迴歸方程式
omega = np.polyfit(x, y ,1)
omega2 = np.polyfit(x, y ,2)
omega3 = np.polyfit(x, y ,3)
omega4 = np.polyfit(x, y ,4)
omega5 = np.polyfit(x, y ,5)

# 測試結果

f = np.poly1d(omega)
f2 = np.poly1d(omega2)
f3 = np.poly1d(omega3)
f4 = np.poly1d(omega4)
f5 = np.poly1d(omega5)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('using polyfitfunction')
plt.grid()
plt.scatter(x, y, marker='x', c='red')
xx = np.linspace(-4, 4, 100)
plt.plot(xx, f(xx), color='green')
plt.plot(xx, f2(xx), color='blue')
plt.plot(xx, f3(xx), color='yellow')
plt.plot(xx, f4(xx), color='purple')
plt.plot(xx, f5(xx), color='orange')
plt.show()
```
